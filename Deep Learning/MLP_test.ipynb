```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pd.read_csv("attributes_std_test.csv", index_col=0)
df.head()

df.head()
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]

print(X.shape)
print(y.shape)
```

```python
X.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report

with open("mlp.pkl", 'rb') as file:
    mlp = pickle.load(file)

mlp.summary()
```

```python
mlp.input_shape
```

```python
classes = np.unique(y)

mlp_predictions = classes[np.argmax(mlp.predict(X), axis=1)]

report = classification_report(y, mlp_predictions)
print(report)

cm = confusion_matrix(y, mlp_predictions)
confusion_matrix_display = ConfusionMatrixDisplay(cm, display_labels=classes)
confusion_matrix_display.plot(cmap="Oranges")
```
