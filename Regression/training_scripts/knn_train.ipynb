```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../attributes_std.csv", index_col=0)
df.head()
```

```python
X = df.loc[:, df.columns != 'Weight']

Y = df["Weight"]

print(np.shape(X))
print(np.shape(Y))
```

```python
# HYPERPARAMETER GRID SEARCH
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV, KFold

from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 15, 21],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'chebyshev'],
    'p': [1, 2]
}

svr = KNeighborsRegressor()

grid_search = GridSearchCV(
    estimator=svr,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    n_jobs=-1
)

# Fit GridSearchCV
grid_search.fit(X, Y)

# Print the best parameters and score
print("Best parameters found: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
```

```python
from sklearn.neighbors import KNeighborsRegressor

svr = KNeighborsRegressor(metric='manhattan', n_neighbors= 11, p= 1, weights='distance')

svr.fit(X, Y)
```

```python
import pickle

# Save models with pickle
with open('../models/knn.pkl', 'wb') as f:
    pickle.dump(svr, f)
```
